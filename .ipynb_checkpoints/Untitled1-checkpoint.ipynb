```python
import pandas as pd
import firebase_admin
from firebase_admin import db
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min
from mpl_toolkits.mplot3d import Axes3D
```

```python
plt.rcParams['figure.figsize'] = (16,9)
plt.style.use('ggplot')
dataframe = pd.read_csv(r"entradasBombillo.csv", sep = ';')
dataframe.head()
```

```python
dataframe.hist()
plt.show()
```

```python
sb.pairplot(dataframe.dropna(), height = 4, vars = ["ALFA", "GAMMA", "EPSILON"], kind= 'scatter')
```

"ALFA", "BETA", "GAMMA", "DELTA", "EPSILON", "ZETA", "ETA", "THETA", "IOTA", "KAPPA", "LAMBDA", "MU", "NU", "XI"

```python
x = np.array(dataframe[["ALFA", "BETA", "GAMMA", "DELTA", "EPSILON", "ZETA", "ETA", "THETA", "IOTA", "KAPPA", "LAMBDA", "MU", "NU", "XI"]])
x.shape
```

```python
kmeans = KMeans(n_clusters=3).fit(x)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
# Predicting the clusters
labels = kmeans.predict(x)
# Getting the cluster centers
C = kmeans.cluster_centers_
colores=['red','green','blue']
asignar=[]
for row in labels:
    asignar.append(colores[row])
 
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x[:, 0], x[:, 1], x[:, 2], c=asignar,s=60)
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c=colores, s=1000)
```
